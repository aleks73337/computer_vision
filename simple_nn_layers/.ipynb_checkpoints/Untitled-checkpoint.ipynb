```python
import PIL.Image as Image
import numpy as np
```

```python
img = Image.open('sample.jpg')
img = img.convert('LA')
```

```python
img
```

```python
def conv2d(img, kernel):
    res_img_w, res_img_h, res_img_c = (img.shape[0] - kernel.shape[0], img.shape[1] - kernel.shape[1], kernel.shape[2])
    result = np.zeros(shape = (res_img_w, res_img_h, res_img_c))
    for c in range(res_img_c):
        for y in range(res_img_h):
            for x in range(res_img_w):
                img_part = img[x : x + kernel.shape[0], y : y + kernel.shape[1]]
                img_res = np.sum(img_part * kernel[:, :, c])
                result[x, y, c] = img_res
    return result
```

```python
kernel = np.random.normal(size = [3, 3, 3])
```

<b> Normalization </b>

```python
img_arr = np.array(img)[:, :, 0] / 255.
```

```python
res_conv = conv2d(img_arr, kernel)
```

```python
Image.fromarray((res_conv * 255).astype(np.uint8))
```

```python
def batch_norm(img):
    mean = np.mean(img)
    std = np.std(img)
    return (img - mean) / std
```

```python
res_bn = batch_norm(res_conv)
```

```python
Image.fromarray((((res_bn + res_bn.min()) / res_bn.max()) * 255).astype(np.uint8))
```

```python
def max_pool(img, kernel_size = (3,3)):
    res_img_w, res_img_h, res_img_c = (img.shape[0] - kernel_size[0], img.shape[1] - kernel_size[1], img.shape[2])
    result = np.zeros(shape = (res_img_w, res_img_h, res_img_c))
    for y in range(res_img_h):
        for x in range(res_img_w):
            img_part = img[x : x + kernel_size[0], y : y + kernel_size[1], :]
            img_res = np.max(img_part, axis = (0, 1))
            result[x, y] = img_res
    return result
```

```python
res_max_pool = max_pool(res_bn, (10, 10))
```

```python
Image.fromarray((((res_max_pool + res_max_pool.min()) / res_max_pool.max()) * 255).astype(np.uint8))
```

```python
flatten = np.reshape(res_bn, -1)
```

```python
flatten.shape
```

```python
def softmax(vec):
    exps = np.exp(vec - np.max(vec))
    return exps / exps.sum(axis=0)
```

```python
res_sm = softmax(flatten)
```

```python
res_sm
```

```python
print(f"Softmax sum == 1? {np.isclose(res_sm.sum(), 1, atol=10e-6)}")
```
